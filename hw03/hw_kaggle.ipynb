# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_curve
```

## Preprocessing

```python
def preprocess_data(data, is_y=True):
    # Использую pandas.get_dummies(), потому что тут понятно как выкидывать лишний признак
    # (флаг drop_first)

    data_train_cat = pd.get_dummies(data[cat_cols], columns=cat_cols, drop_first=True)

    # Preparing scalers' parameters for each numerical feature

    # Попробовать также Robust scaler
    scaler = StandardScaler()
    data_train_num = pd.DataFrame(scaler.fit_transform(data[num_cols]), columns=num_cols)

    X = pd.concat([
            pd.DataFrame(data_train_num, columns=num_cols),
            data_train_cat
    ], axis=1)

    if not is_y:
        return X
    
    y = data[target_col].values

    return X, y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    *preprocess_data(data), test_size=0.2, random_state=42)
```

```python
log_model = LogisticRegressionCV(
    Cs = [0.001, 0.01, 0.1, 1, 10, 100],
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    refit=True
)
```

```python
log_model.fit(X_train, y_train)

y_train_pred = log_model.predict_proba(X_train)[:,1]
y_test_pred = log_model.predict_proba(X_test)[:,1]
```

```python
train_auc = roc_auc_score(y_train, y_train_pred)
test_auc = roc_auc_score(y_test, y_test_pred)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_pred)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_pred)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
import catboost
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[feature_cols], data[target_col], 
                                                       train_size=0.8, 
                                                       random_state=42)
```

```python
boosting_model = catboost.CatBoostClassifier(n_estimators=300,
                                             cat_features=cat_cols,
                                             silent=True,
                                            custom_metric=['AUC', 'AUC:hints=skip_train~false'])
# max_depth
boosting_model.grid_search(
    {
        #'n_estimators' : np.arange(100, 225, 25),
        #'max_depth' : np.arange(7, 10),
        'l2_leaf_reg' : [0.001, 0.01, 0.1, 1, 10, 100]
    }, 
    X_train, y_train, 
    cv=5, refit=True, 
    train_size=0.8
)

boosting_model.get_best_iteration()

y_train_predicted = boosting_model.predict_proba(X_train)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_test)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```
